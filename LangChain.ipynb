## GENAI GEMINI

```python
from google.colab import userdata
GEMINI=userdata.get('Gemini_API')

import google.generativeai as genai
genai.configure(api_key = GEMINI)

for models in genai.list_models():
  print(models)
```

```python
# to filter the list of mdoels which can help me generate content
for models in genai.list_models():
    if "generateContent" in models.supported_generation_methods:
        #print(models) # this will give details about the mdoels
        print(models.name)
```

```python
# Suppose I want to use any of the model to generate content
Gemini_model = genai.GenerativeModel("models/gemini-2.5-flash-lite")
```

## 2_basic_conversation

```python
## M-1

# 📝 Initialize chat history as a list of strings
chat_history = []

# 💬 Chat loop
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break

    # Add user input to chat history
    chat_history.append(f"User: {user_input}")

    # Create prompt by joining all history
    prompt = "\n".join(chat_history)

    # Get response from Gemini
    response = Gemini_model.generate_content(prompt)
    ai_response = response.text

    # Show and save AI response
    print("AI:", ai_response)
    chat_history.append(f"AI: {ai_response}")
```

```python
### M-2
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"

gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=GEMINI)

response = gemini.chat.completions.create(model="gemini-2.5-flash-lite", messages=[{"role": "user", "content": "Tell me a fun fact"}])

response.choices[0].message.content
```
