```python
# Suppose I want to use any of the model to generate content
Gemini_model = genai.GenerativeModel("models/gemini-exp-1206")
```

## 2_basic_conversation

```python
######### GPT model call #########

from langchain.schema import AIMessage, HumanMessage, SystemMessage

chat_history = [] # use a list to store message
system_message = SystemMessage(content="You are a helpful AI assistant.")
chat_history.append(system_message)

# looping chat with AI
while True:
    query = input("You: ")
    if query == "exit":
        break
    chat_history.append(HumanMessage(content=query))

    response = GPT_model.invoke(chat_history)
    print("AI: ", response.content)
    chat_history.append(response)

print(chat_history)
```

```python
# 📝 Initialize chat history as a list of strings
chat_history = []

# 💬 Chat loop
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break

    # Add user input to chat history
    chat_history.append(f"User: {user_input}")

    # Create prompt by joining all history
    prompt = "\n".join(chat_history)

    # Get response from Gemini
    response = Gemini_model.generate_content(prompt)
    ai_response = response.text

    # Show and save AI response
    print("AI:", ai_response)
    chat_history.append(f"AI: {ai_response}")
```
